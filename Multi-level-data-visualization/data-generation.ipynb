# Generate multi-level dataset

```python
import pandas as pd 
import random
import configparser
from tqdm.auto import tqdm
```

## Parameters

```python
config = configparser.ConfigParser()
sectionName = 'Generation'
config.add_section(sectionName)
config[sectionName]['Total'] = '1000'
config[sectionName]['Levels'] = '5'
config[sectionName]['LevelSegments'] = '{}'.format((3, 10, 20, 80, 100))
config[sectionName]['ValueRange'] = '{}'.format((3, 10))
config.write(open('setting.ini', 'w'))
```

## Generation

```python
total = int(config[sectionName]['Total'])
levels = int(config[sectionName]['Levels'])
rnd_min, rnd_max = eval(config[sectionName]['ValueRange'])
segments = eval(config[sectionName]['LevelSegments'])

columns = [f'Level_{e}' for e in range(levels)]
# columns.append('Value')

data = pd.DataFrame(columns=columns)
data['Value'] = [random.randint(rnd_min, rnd_max) for e in range(total)]

def rnd_feed(level):
    num = segments[level]
    candidates = [f'No-{e}-Lvl-{level}' for e in range(num)]
    return random.choice(candidates)

for j, col in tqdm(enumerate(columns)):
    data[col] = j
    data[col] = data[col].map(lambda level: rnd_feed(level))

data
```
